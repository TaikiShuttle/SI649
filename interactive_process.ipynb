```python
import pandas as pd

df = pd.read_csv('insurance.txt', sep = '\t')
```

```python
df = df.iloc[1: 52,:]
df['Estimate (%)'] = df['Estimate (%)'].astype(float)
df.to_csv('insurance.csv', index = False)
```

```python
df2 = pd.read_csv('income.txt', sep = '\t')

df2 = df2.iloc[1: 52,:]
df2['Estimate ($)'] = df2['Estimate ($)'].apply(lambda x: x.replace(',', '')).astype(int)
df2.to_csv('income.csv', index = False)
```
